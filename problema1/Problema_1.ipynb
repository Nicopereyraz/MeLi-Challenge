```python
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
from datetime import datetime
```

```python
df = pd.read_csv('ofertas_relampago.csv')
```

```python
df.head()
```

```python
#Generacion de variable Duracion
a = [datetime.fromisoformat(x) for x in df['OFFER_START_DTTM']]
b = [datetime.fromisoformat(x) for x in df['OFFER_FINISH_DTTM']]
df['DURACION'] = [(b[x]-a[x]).total_seconds() for x in range(len(a))]
```

```python
df[df['DURACION']==0]
```

```python
#Generacion de variable precio unitario
df['PRECIO_UNITARIO'] = df['SOLD_AMOUNT']/df['SOLD_QUANTITY']
```

```python
df.head()
```

```python
#defining the library
profile = ProfileReport(df)
# exporting your report as a html file
profile.to_file(output_file='report.html')
```

```python
df.to_csv('dataset_refinado.csv')
```
